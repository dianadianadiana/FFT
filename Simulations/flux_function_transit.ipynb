```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
from pylab import *
from scipy.optimize import curve_fit
```

```python
def flux(t, x):
    """ Takes a time array and a parameter array to output a flux array
    
    t: time array
    x: parameter array == [delta, t0, period, d, sigma, time_int, f0]
        x[0] = delta: change in flux
        x[1] = t0: time of transit (hours)
        x[2] = period: orbital period (hours)
        x[3] = T
        x[4] = tau
        x[5] = sigma: amount of scatter
        x[6] = time_int: time interval of receiving data in hours
        x[7] = f0: the flux when there is no transit
    """
    delta, t0, period, T, tau, sigma, time_int, f0 = x
    orbital_phase = (t-t0)/period # orbital_phase = (t-t0)/period ##phase = 0 at t0
    m = np.modf(orbital_phase)[1] # takes the integer part of the array
    t_m = t0 + m*period# t(m) = t0 + m(t) * period

    # when the period is not nicely divisble by the time interval
    if (period/time_int) % 1 != 0:
        
        # b: array with indicies where m changes
        # ex. m[16] = 0 and m[17] = 1, then 17 would be stored in array b
        i = 1
        b = np.empty(0)
        while i < len(m):
            if m[i-1] != m[i]:
                b = np.append(b, i)
            i += 1

        # c: array (same len as b) with values to correct t-t_m
        c = np.empty(0, dtype = float)
        for i in b:
            c = np.append(c, t[i] - t_m[i])
            
        # double loop that corrects t-t_m         
        i = 0
        while i < len(c):
            j = b[i]
            if i == len(c) - 1:
                upper = len(t-t_m)
            else:
                upper = b[i+1]
            while j < upper:
                t_m[j] = t_m[j] + c[i]
                j += 1
            i += 1
    
    dur = T + tau #duration of transit
    transit_arr = np.empty(0)
    
    i = 1
    while i < len(t_m):
        if t[i]-t_m[i] == 0.0:
            transit_arr = np.append(transit_arr, i)
        i += 1
    
    j = 0
    while j < len(transit_arr):
        index = transit_arr[j]
        t_m[index-dur:index] = t_m[index-dur:index] + period
        j += 1
    
    flux_mod = np.piecewise(t-t_m, 
                            [np.abs(t-t_m) <= (T/2 - tau/2), 
                             np.abs(t-t_m) > T/2 - tau/2, 
                             np.abs(t-t_m) >= T/2 + tau/2],
                            [f0 - delta, 
                             lambda t: f0 - delta + (delta/tau)*(np.abs(t-t0) - T/2 + tau/2),
                             f0])
    
    # (b - a) * random_sample() + a for random floats in interval [a,b)
    scatter = sigma * ((np.random.random_sample(len(t))) * 2*f0 - f0)                                            
    flux_mod = flux_mod + scatter

    return flux_mod
```

```python
### Variables

delta = .1 # depth    # (Rplanet / Rstar)**2
t0 = 0.0 # time of transit in hours
period = 35.549379456876345 # orbital period in hours
#d = period**(-2/3) # duration of transit in hours -- 
        # scales with P^(-2/3)
    
# look at pg 6 http://arxiv.org/pdf/0805.0238v2.pdf
T = 4.7
tau = 0.1### need to fix flux funtion if tau is 0
sigma = 0.0 #scatter
time_int = 30/60. # time interval of receiving data, in hours
f0 = 1.0 # base flux
x = [delta, t0, period, T, tau, sigma, time_int, f0]

print("Period: " + str(period) + " hours")
print("Duration of Transit: " + str(T+tau) + " hour(s)")
print("Time Interval of collecting data: " + str(time_int) + " hours")
```

```python
### Setting up a Time Array
### --- time scatter --- has gaps ---

a, b = 0, 1000 #hours #a,b : min and max of time
#N: number of data points
N = np.round((b-a) / time_int) #float
print("Number of data points: " + str(N))
t = np.linspace(a, b, N, endpoint = False) #hours
print("Number of Periods: " +str((b-a)/period))

time_sigma = 0.0/60 # in hours
time_scatter = time_sigma * (np.random.random_sample(len(t)))
t = t + time_scatter

t_random =  np.around(np.random.random(len(t))) # create array randomly with 0's and 1's
count_one = np.float(np.count_nonzero(t_random)) # count the number of "1's" in the array
ratio = 1.0 # the ratio of how many real data points we have to how many there are total
while count_one/len(t_random) < ratio:
    loc_zero = np.transpose(np.argwhere(t_random == 0.0))[0] # arr with indexes of where the 0's are
   
    # create a random bool array with the same length as loc_zero
    # true: data point, false: gap point
    loc_zero_bool = np.array(np.around(np.random.random(len(loc_zero))),dtype=bool) 
    
    i = 0 
    while i < len(loc_zero):
        index = loc_zero[i] 
        if loc_zero_bool[i]:
            t_random[index] = 1.0
        i+=1
    count_one = np.float(np.count_nonzero(t_random)) # recount all the 1's now

t_bool = array(t_random, dtype=bool) # make a boolean array -- True is a data point, False is a gap
t_bool[:2], t_bool[len(t_bool) - 3:] = True, True # make sure few end points are data points

t = t[t_bool] # keep only times where it is a data point
print("There are " + str(len(t)) + " data points and there are " + str(N-len(t)) + " missing points")

#distance between points
d_pts = (b - a) / N
#Nyquist frequency # https://en.wikipedia.org/wiki/Nyquist_frequency
nyquist_freq = .5 / d_pts 
print("Nyquist frequency: " + str(nyquist_freq))
```

```python
#Determine the appropriate cadence to which the data should be interpolated
# Get the GAP array
num_time = len(t) # the original number of data time points (w gaps)
f = flux(t,x) # the flux values (w gaps)

delta_t = t[1:num_time-1]-t[0:num_time-2] # has a length = len(t) - 2 (bc it ignores endpoints)
cadence = np.mean(delta_t) # what the time interval between points with gaps currently is
#Fill gaps in the time series (according to 2012 thesis method)
gap = delta_t/cadence # has a length = len(t) - 2 (bc it ignores endpoints)
gap = np.around(gap, decimals = 0)

gap_cut = 1.1   #Time differences greater than gap_cut*cadence are gaps    
gap_loc = [i for i in np.arange(len(gap)) if (gap[i] > gap_cut)]
num_gap = len(gap_loc) # the number of gaps present 
biggest_gap = np.amax(gap)

print("num_gap: " +str(num_gap))
print("biggest gap: " + str(biggest_gap))
```

```python
### Create a new time and flux array to account for the gaps

if num_gap == 0:
    num_cad = sum(gap) + 2
else:
    num_cad = sum(gap) + 1 # the number of data points
time_cad = np.arange(num_cad)*cadence + a # a: starting value of t ("t min")
flux_cad = np.arange(num_cad, dtype = np.float) 

print("***Filling in the Gaps***")
print("number of data points now: " +str(num_cad) + " vs before: " + str(num_time))

oldflux_st = 0 # this is an index
newflux_st = 0 # this is an index
n = 0
while n < num_gap - 1:
    oldflux_ed = gap_loc[n]
    gap_sz     = gap[gap_loc[n]]
    newflux_ed = newflux_st + (oldflux_ed-oldflux_st)
    
    flux_cad[newflux_st:newflux_ed] = f[oldflux_st:oldflux_ed]
    
    gap_fill = (np.zeros(gap_sz - 1, dtype = np.float) + 1.)
    # times the mean of the two values before it
    gap_fill[0] = gap_fill[0] * np.mean(f[oldflux_ed-1:oldflux_ed +1]) 
    if gap_sz > 2:
        gap_fill[1] = gap_fill[1] * np.mean([f[oldflux_ed],gap_fill[0]])
        j = 2
        while j < len(gap_fill):
            gap_fill[j] = gap_fill[j] * np.mean([gap_fill[j-2],gap_fill[j-1]])
            j += 1
            
    flux_cad[newflux_ed+1:newflux_ed+gap_sz] = gap_fill
    flux_cad[newflux_ed] = np.mean([flux_cad[newflux_ed-1], flux_cad[newflux_ed+1]])
    
    oldflux_st = oldflux_ed + 1
    newflux_st = newflux_ed + gap_sz

    n+=1 

flux_cad[newflux_st:] = f[oldflux_st:] #account for last part where there is no gap after
```

```python
#plot the lightcurves

#plot lightcurve all data points
fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(2,1,1)
ax.scatter(time_cad,flux_cad)
ax.plot(time_cad,flux_cad)

plt.title("Lightcurve (all data points)")
plt.xlabel("Time (Hours)")
plt.ylabel("Flux")
plt.xlim([a,b])
plt.ylim([1 - 1.5*delta, 1. + .5*delta])
plt.show()
```

```python
### FFT part

# oversampling
N_log = np.log2(N) # 2 ** N_log = N
exp = np.round(N_log)
if exp < N_log:
    exp += 1 #compensate for if N_log was rounded down

extra_fact = 5
newN = 2**(exp + extra_fact)

n = newN/N
n = np.round(n)

diff = newN - N
mean = np.mean(f)
voidf = np.zeros(diff) + mean
newf = np.append(flux_cad, voidf)

norm_fact = 2.0 / newN # normalization factor 
f_flux = fft(newf) * norm_fact
    
freq = fftfreq(len(newf))
freq_fact = 1.0 / d_pts #frequency factor 

postivefreq = freq > 0 # take only positive values
freq, f_flux = freq[postivefreq], f_flux[postivefreq]

power = np.abs(f_flux)

conv_hr_day = 24. #conversion factor from cycles/hour to cycles/day
constant = freq_fact*conv_hr_day
#because of oversampling, need to remove the first few terms, 
#ex if N=3000 and the over sampling size is 100,000, then the first 30 points should not be considered
#also means that each cluster of higher amplitudes are in groups of 30 points
plt.plot(freq[n:]*constant, power[n:],'b+')

plt.title("FFT")
plt.xlabel("Frequency (cycles/day)")
plt.ylabel("Amplitude")
# plt.xlim([1,10])
plt.xlim([0,10])
plt.show()
```

```python
# GOAL: find all the values in the power array that are considered to be peaks
#       and all the values in the power array that are considered to be harmonics
    
def constraint_index_finder(constraint, x, y, poly = 0, loops = 10):
    """ 
    Takes two arrays (freq and power) and then computes the line of best fit/mean value (using poly)
    The value is then multiplied with the constraint, and any value in y (power) that is above
        constraint * mean value is stored to an array --> That array is then returned
    Return value: 
        val_indexes -- array with indexes where y[indexes] output values greater than 
                    constraint * mean value
    
    constraint: the number times the best fit/mean of y
    x: x values -- freq
    y: y values -- power
    poly: the power of the best line fit
    loops: number of times the mean will be getting recalculated
    
    """
    # raise error if x and y don't have the same length
    if len(x) != len(y):
        raise Exception('x and y needs to have the same length.')
                        
    # val_indexes: stores all the indexes in y that have 
    # a power value greater constraint * value of best line fit
    val_indexes= np.empty(0) 
    j = 0
    while j < loops:
        z = np.polyfit(x, y, poly)
        if j > 0 and z[0] == val:
            break # to leave the loop early if the same value is brought up
        val = z[0]
        i = 0
        while i < len(y):
            if y[i] >= constraint * val:
                val_indexes = np.append(val_indexes, i)
            i += 1
        j += 1
    val_indexes = val_indexes.astype(int)
    return val_indexes
```

```python
def poly_maker(x, y, func, poly = 1):
    #z1 = z[0]*t**5 + z[1]*t**4 + z[2]*t**3 + z[3]*t**2 + z[4]*t + z[5]
    z = np.polyfit(x, y, poly)
    i = 0
    znew = np.zeros(len(func)) + z[poly]
    while i < poly:
        znew = znew + z[i]*func**(poly-i)
        i += 1
    return znew
# coefficients = polyfit(x, y, 6)
# polynomial = poly1d(coefficients)
# xs = arange(-2.2, 2.6, 0.1)
# ys = polynomial(xs)
```

```python
# Your data set has 10 equidistant data points. 
# Let's pick the three data points (110, 2391), (350, 786), (590, 263) 
# for use ― these have the greatest possible fixed distance (240) in the independent coordinate.
# So, y_0 = 2391, y_1 = 786, y_2 = 263, k = 110, s = 240. 
# Then A = 10.20055, B = 2380.799, C = 0.3258567, 
# A′ = 10.20055, B′ = 3980.329, C′ = 0.9953388
# http://stackoverflow.com/questions/3938042/fitting-exponential-decay-with-no-initial-guessing

def isNaN(num):
    return num != num

def parameter_finder(arr, arr1):
    index = .25 * (len(arr))
    i0, i1, i2 = index, 2*index, 3*index
    width = 30
    x0, x1, x2 = arr[i0], arr[i1], arr[i2]
    y0, y1, y2 = np.mean(arr1[i0-width:i0+width]), np.mean(arr1[i1-width:i1+width]), np.mean(arr1[i2-width:i2+width])
    k, s = x0, x1 - x0
    a = (y0*y2 - y1**2)/(y0 + y2 - 2*y1)
    b = (y1 - y0)**2/(y0 + y2 - 2*y1)
    c = (y2 - y1)/(y1 - y0)
    A = a
    if isNaN(c**(1./s)):
        B, C = 1, 1
    else:
        C = c**(1./s)
        B = b/(C**k)    
    return np.array([A, B, C])

def exp_decay(x, A, B, C):
    return A + B*C**x
```

```python
# cluster algorithm
# goal: find out where the clusters are, and keep only the max of each cluster

def cluster(index_arr, arr, peak_width = .13):
    """
        Takes an array of values (indexes) and clusters them (based on the width)
        Parameters:
            index_arr -- array of indexes (sub array from a bigger array)
            arr -- array of values (it is a sub array from a bigger array)
            peak_width -- how wide is the peak
        Returns:
            clusters -- a list of lists. each element of clusters is one cluster
    """
    clusters = []
    i = 0
    while i < len(index_arr):
        temp_arr = [index_arr[i]]
        j = i+1
        while j < len(index_arr) and (np.abs(arr[temp_arr[0]] - arr[index_arr[j]]) <= peak_width):
            temp_arr.append(index_arr[j])
            j+=1
        clusters.append(temp_arr)
        i=j
    return clusters

def cluster_max(clusters, y):
    """ 
        Takes a list of lists (clusters) and finds the index of where each max value of a cluster is
        Parameters:
            clusters -- a list of lists
            y -- the bigger array (with length newN)
        Returns:
            max_indexes -- an array of the indexes of where the max of each cluster is
    """
    max_indexes = np.empty(0)
    for elem in clusters:
        max_indexes = np.append(max_indexes, elem[y[elem].argmax()])
    max_indexes = max_indexes.astype(int)
    return max_indexes

# print(cluster(val_indexes, freq*constant, peak_width))
# print(cluster(val_indexes1, freq*constant, peak_width))

peak_indexes = cluster_max(cluster(val_indexes, freq*constant,peak_width), power_rel)    
harmonics_indexes = cluster_max(cluster(val_indexes1, freq*constant,peak_width), power_rel)  

print("Indexes of the max values of each cluster above " + 
      str(peak_constraint) + ": " + str(peak_indexes))
print("Indexes of the max values of each cluster above "+ 
      str(harmonics_constraint) + ": " + str(harmonics_indexes))
```

```python
def peak_verifier(index_arr, arr1, arr2, n, peak_width):
    """
        The purpose of this function is to recognize if two points are on the same peak, 
            and if there are two poitns on the same peak, then the point that is the true
            max of the peak will be chosen, whereas the other one will be discarded.
        Parameters:
            index_arr: this is the peak_indexes or harmonics_indexes
            arr1: freq*constant
            arr2: power
            n: 2*n = the minumum number of data points for a peak
            peak_width: the width of the peak
        Returns:
            the index_arr with the indexes of the true peak maxes
    """
    k = 0
    delete_arr = []
    while k < len(index_arr) - 1:
        curr_index = index_arr[k]
        next_index = index_arr[k+1]

        if np.abs(arr1[curr_index]-arr1[next_index]) <= peak_width:

            curr_lower, curr_upper = curr_index - n, curr_index + n
            next_lower, next_upper = next_index - n, next_index + n

            #curr_maxpow and next_maxpow should be the same
            curr_maxpow = np.amax(arr2[curr_lower:curr_upper])
            next_maxpow = np.amax(arr2[next_lower:next_upper])
            maxpow = np.amax([curr_maxpow,next_maxpow])

            if arr2[curr_index] == maxpow:
                delete_arr.append(k+1)
            else:
                delete_arr.append(k)  
        k+=1
    return np.delete(index_arr, delete_arr)
    
peak_indexes = peak_verifier(peak_indexes, freq*constant, power_rel, n, peak_width)
harmonics_indexes = peak_verifier(harmonics_indexes, freq*constant, power_rel, n, peak_width)

print(peak_indexes)
print(harmonics_indexes)
```

```python
def limit_applier(arr, lower_limit = 1.0, upper_limit = 10.0):
    """
    Takes an array and a lower and upper limit and returns an array of the indices that 
        are below the lower limit or higher than the upper limit -- it is the delete_arr
    """
    delete_arr = np.empty(0)
    i = 0
    while i < len(arr):
        if arr[i] < lower_limit:
            delete_arr = np.append(delete_arr, i)
        elif arr[i] > upper_limit:
            delete_arr = np.append(delete_arr, np.arange(i, len(arr)))
            break 
        i += 1
    delete_arr = delete_arr.astype(int)
    return delete_arr
```
